```python
# Erstellen Sie eine Funktion, die die Anzahl der Siege, Unentschieden und Gesamtspiele für eine gegebene Gruppe von Spielen berechnet
def calculate_stats(group):
    continents = sorted(group.name)
    total_games = len(group)
    continent1_wins = len(group[group["winning_continent"] == continents[0]])
    continent2_wins = len(group[group["winning_continent"] == continents[1]])
    draws = len(group[group["winning_continent"] == "Unentschieden"])
    return pd.Series([continents[0], continents[1], continent1_wins, continent2_wins, draws, total_games], index=df_help.columns)

# Create a new column with the sorted tuple of "home_continent" and "away_continent"
df_results["continent_pair"] = df_results[["home_continent", "away_continent"]].apply(lambda x: tuple(sorted(x)), axis=1)

# Define the columns for the new DataFrame
df_help = pd.DataFrame(columns=["Continent1", "Continent2", "Continent1 Wins", "Continent2 Wins", "Draws", "Total Games"])

# Apply the calculate_stats function to each group of games grouped by the sorted continent pair
df_help = df_results.groupby("continent_pair").apply(calculate_stats).reset_index(drop=True)

#df_help.to_csv("input/df_help.csv", encoding="utf-8", index=False)

df_help
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pivot = df_help.pivot_table(index='Continent1', columns='Continent2', values='Continent1 Wins', aggfunc='sum')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap="YlGnBu")
plt.title('Heatmap of Wins by Continent')
plt.show()
```

```python
df = df_help.copy()

# Aggregating the data by continent
def aggregate_data(continent, df):
    continent_data = df[(df['Continent1'] == continent) | (df['Continent2'] == continent)]
    continent_data = continent_data[continent_data['Continent1'] != continent_data['Continent2']]

    wins = continent_data['Continent1 Wins'].sum()
    losses = continent_data['Continent2 Wins'].sum()
    draws = continent_data['Draws'].sum()
    
    return wins, losses, draws

continent_list = df['Continent1'].unique().tolist()
continent_list.extend(df['Continent2'].unique().tolist())

continents = list(set(continent_list))
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for i, continent in enumerate(continents):
    ax = axs[i // 3, i % 3]
    wins, losses, draws = aggregate_data(continent, df)
    labels = ['Wins', 'Losses', 'Draws']
    sizes = [wins, losses, draws]
    colors = ['#4CAF50', '#FF5733', '#FFC300']
    explode = (0.1, 0, 0)  # explode the 1st slice (Wins)
    ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
    ax.set_title(f'{continent} - Distribution of Results')

plt.tight_layout()
plt.show()
```
