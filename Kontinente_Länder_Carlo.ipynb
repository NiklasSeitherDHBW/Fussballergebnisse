```python
# Anzahl der Heimspiele und Heimtore pro Land berechnen
home_stats = df_results.groupby('home_team').agg(
    home_goals=('home_score', 'sum'),
    home_games=('home_score', 'count')
).reset_index()
home_stats.columns = ['team', 'home_goals', 'home_games']

# Anzahl der Ausw√§rtsspiele und Ausw√§rtstore pro Land berechnen
away_stats = df_results.groupby('away_team').agg(
    away_goals=('away_score', 'sum'),
    away_games=('away_score', 'count')
).reset_index()
away_stats.columns = ['team', 'away_goals', 'away_games']

# Zusammenf√ºhren der Heim- und Ausw√§rtsstatistiken
total_stats = pd.merge(home_stats, away_stats, on='team', how='outer').fillna(0)

# Berechnung der Gesamtanzahl der Tore und Spiele
total_stats['total_goals'] = total_stats['home_goals'] + total_stats['away_goals']
total_stats['total_games'] = total_stats['home_games'] + total_stats['away_games']

# Berechnung der durchschnittlichen Tore pro Spiel
total_stats['avg_goals_per_game'] = total_stats['total_goals'] / total_stats['total_games']

# Sortieren nach der durchschnittlichen Tore pro Spiel absteigend
total_stats = total_stats.sort_values(by='avg_goals_per_game', ascending=False)

# Rundung der durchschnittlichen Tore pro Spiel auf 2 Nachkommastellen
total_stats['avg_goals_per_game'] = total_stats['avg_goals_per_game'].round(2)

# Das Ergebnis anzeigen
print(total_stats)
```

```python
df_shootouts['home_continent'] = df_shootouts['home_team'].map(continents)
df_shootouts['away_continent'] = df_shootouts['away_team'].map(continents)
```

```python
df_shootouts["home_continent"] = df_shootouts["home_continent"].fillna("Country not in Fifa Ranking")
df_shootouts["away_continent"] = df_shootouts["away_continent"].fillna("Country not in Fifa Ranking")
```

```python
no_country_in_fifa_ranking_both = df_shootouts[(df_shootouts["home_continent"] == "Country not in Fifa Ranking") | (df_shootouts["away_continent"] == "Country not in Fifa Ranking")]

df_shootouts = df_shootouts.drop(no_country_in_fifa_ranking_both.index)
```

```python
# Z√§hle die Teilnahmen an Elfmeterschie√üen pro Land (Heim)
home_shootouts = df_shootouts.groupby('home_team').apply(lambda x: (x['home_team'] != x['winner']).sum()).reset_index(name='home_shootouts')

# Z√§hle die Teilnahmen an Elfmeterschie√üen pro Land (Ausw√§rts)
away_shootouts = df_shootouts.groupby('away_team').apply(lambda x: (x['away_team'] != x['winner']).sum()).reset_index(name='away_shootouts')

# Zusammenf√ºhren der Ergebnisse
shootouts_total = pd.merge(home_shootouts, away_shootouts, left_on='home_team', right_on='away_team', how='outer').fillna(0)

# Summieren der Heim- und Ausw√§rtsschie√üereien f√ºr die Gesamtanzahl
shootouts_total['total_shootouts'] = shootouts_total['home_shootouts'] + shootouts_total['away_shootouts']

# Berechnung der Anzahl der gewonnenen Elfmeterschie√üen (Heim)
home_wins = df_shootouts[df_shootouts['home_team'] == df_shootouts['winner']].groupby('home_team').size().reset_index(name='home_wins')

# Berechnung der Anzahl der gewonnenen Elfmeterschie√üen (Ausw√§rts)
away_wins = df_shootouts[df_shootouts['away_team'] == df_shootouts['winner']].groupby('away_team').size().reset_index(name='away_wins')

# Zusammenf√ºhren der Ergebnisse
wins_total = pd.merge(home_wins, away_wins, left_on='home_team', right_on='away_team', how='outer').fillna(0)

# Merge wins_total mit shootouts_total
shootouts_total = pd.merge(shootouts_total, wins_total, left_on='home_team', right_on='home_team', how='outer').fillna(0)

# Summieren der Heim- und Ausw√§rtssiege f√ºr die Gesamtanzahl
shootouts_total['total_wins'] = shootouts_total['home_wins'] + shootouts_total['away_wins']

# Berechnung der Win-Rate
shootouts_total['home_win_rate'] = shootouts_total['home_wins'] / shootouts_total['home_shootouts']
shootouts_total['away_win_rate'] = shootouts_total['away_wins'] / shootouts_total['away_shootouts']
shootouts_total['total_win_rate'] = shootouts_total['total_wins'] / shootouts_total['total_shootouts']

# Nach total_shootouts absteigend sortieren
shootouts_total = shootouts_total.sort_values(by='total_shootouts', ascending=False)

# Tabelle f√ºr Heim-Teilnahmen und Win-Rate
home_table = shootouts_total[['home_team', 'home_shootouts', 'home_wins', 'home_win_rate']].copy()
home_table.columns = ['Team', 'Shootouts (Home)', 'Wins (Home)', 'Win Rate (Home)']

# Tabelle f√ºr Ausw√§rts-Teilnahmen und Win-Rate
away_table = shootouts_total[['home_team', 'away_shootouts', 'away_wins', 'away_win_rate']].copy()
away_table.columns = ['Team', 'Shootouts (Away)', 'Wins (Away)', 'Win Rate (Away)']

# Tabelle f√ºr Gesamt-Teilnahmen und Win-Rate
total_table = shootouts_total[['home_team', 'total_shootouts', 'total_wins', 'total_win_rate']].copy()
total_table.columns = ['Team', 'Total Shootouts', 'Total Wins', 'Total Win Rate']

# Das Ergebnis anzeigen
print("Home Shootouts Table:")
print(home_table)
print("\nAway Shootouts Table:")
print(away_table)
print("\nTotal Shootouts Table:")
print(total_table)
home_table.to_csv('home_shoot', index=False)
away_table.to_csv('away_shoot', index=False)
total_table.to_csv('total_shoot', index=False)
```
